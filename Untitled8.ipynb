```python
# Summing up energy consumption by source for a specific year (e.g., 2020)
year_data = df[df['year'] == 2020]
energy_sources = ['coal_consumption', 'oil_consumption', 'gas_consumption', 'solar_consumption', 'wind_consumption', 'hydro_consumption']
total_energy = year_data[energy_sources].sum()

plt.figure(figsize=(8,8))
plt.pie(total_energy, labels=energy_sources, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title('Energy Consumption Share by Source (2020)')
plt.show()
```

```python
# Aggregating data by country and year, selecting top 5 countries
top_countries = df.groupby('country')['primary_energy_consumption'].sum().nlargest(5).index
top_data = df[df['country'].isin(top_countries)]

plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
for country in top_countries:
    country_data = top_data[top_data['country'] == country]
    plt.fill_between(country_data['year'], country_data['primary_energy_consumption'], label=country, alpha=0.6)

plt.title('Top 5 Countries\' Energy Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Energy Consumption (TWh)')
plt.legend()
plt.show()
```
