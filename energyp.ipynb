```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")

# Load data (replace with the correct file path if running locally)
data = pd.read_csv("C:\\Users\\ashmp\\OneDrive\\Desktop\\energy-consumption-2020-1.csv")

# Plot: Distribution of Building Types
plt.figure(figsize=(10,6))
plt.scatter(data['Longitude'], data['Latitude'], alpha=0.7, c=data['Electricity Usage'], cmap='coolwarm')
plt.colorbar(label='Electricity Usage (kWh)')
plt.title('Geographic Distribution of Sites', fontsize=16)
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Latitude', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=data, x='Department', y='Electricity Usage', estimator=sum, palette="viridis")
plt.title('Total Electricity Usage by Department', fontsize=16)
plt.ylabel('Electricity Usage (kWh)', fontsize=14)
plt.xlabel('Department', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=data, x='Department', y='Energy Use Intensity', estimator='mean', palette='magma')
plt.title('Average Energy Use Intensity by Department', fontsize=16)
plt.ylabel('Energy Use Intensity', fontsize=14)
plt.xlabel('Department', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Correlation matrix
corr_matrix = data[['Electricity Usage', 'Peak Electric Demand', 'Building Area', 'Natural Gas Usage', 'Energy Use Intensity']].corr()

# Plot: Heatmap of Correlations
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables', fontsize=16)
plt.show()
```

```python
# Plot: Violin plot of Peak Electric Demand by Department
plt.figure(figsize=(12,6))
sns.violinplot(data=data, x='Department', y='Peak Electric Demand', palette='Set2')
plt.title('Peak Electric Demand by Department', fontsize=16)
plt.xlabel('Department', fontsize=14)
plt.ylabel('Peak Electric Demand (kW)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Sample dataset for energy consumption by sectors over time
data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Residential': [120, 150, 170, 200, 230],
    'Commercial': [100, 120, 140, 160, 180],
    'Industrial': [180, 200, 230, 250, 270]
}

df = pd.DataFrame(data)

# Plotting stacked area chart
plt.figure(figsize=(10,6))
plt.stackplot(df['Year'], df['Residential'], df['Commercial'], df['Industrial'],
              labels=['Residential', 'Commercial', 'Industrial'], colors=['skyblue', 'lightgreen', 'lightcoral'])

plt.title('Energy Consumption by Sector Over Time')
plt.xlabel('Year')
plt.ylabel('Energy Consumption (in TWh)')
plt.legend(loc='upper left')

plt.grid(True)
plt.show()
```

```python
# Sample data for energy consumption by sectors
data = {
    'Sectors': ['Residential', 'Commercial', 'Industrial', 'Transport'],
    'Energy_Consumption': [30, 25, 35, 10]  # in %
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting pie chart
plt.figure(figsize=(7,7))
plt.pie(df['Energy_Consumption'], labels=df['Sectors'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral', 'gold'])

plt.title('Energy Consumption by Sector (Percentage)')
plt.show()
```

```python
# Sample dataset for energy consumption by sector over time
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Residential': [1200, 1300, 1350, 1400, 1500],
    'Commercial': [800, 900, 950, 1000, 1100],
    'Industrial': [2000, 2000, 2000, 2100, 2200]
}

df_sectors = pd.DataFrame(data)

# Plotting stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_sectors['Year'], df_sectors['Residential'], color='lightblue', label='Residential')
plt.bar(df_sectors['Year'], df_sectors['Commercial'], bottom=df_sectors['Residential'], color='lightgreen', label='Commercial')
plt.bar(df_sectors['Year'], df_sectors['Industrial'], bottom=df_sectors['Residential'] + df_sectors['Commercial'], color='lightcoral', label='Industrial')

plt.title('Stacked Energy Consumption by Sector Over Time')
plt.xlabel('Year')
plt.ylabel('Energy Consumption (in TWh)')
plt.legend()
plt.xticks(df_sectors['Year'])
plt.grid(axis='y')
plt.show()
```

```python
# Sample dataset for energy consumption by country over time
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'USA': [4000, 4100, 4200, 4300, 4400],
    'China': [6000, 6200, 6400, 6500, 6700],
    'India': [3000, 3200, 3300, 3400, 3600],
}

df_countries = pd.DataFrame(data)

# Plotting multiple line charts
plt.figure(figsize=(10, 6))
for country in df_countries.columns[1:]:
    plt.plot(df_countries['Year'], df_countries[country], marker='o', label=country)

plt.title('Energy Consumption by Country Over Time')
plt.xlabel('Year')
plt.ylabel('Energy Consumption (in TWh)')
plt.xticks(df_countries['Year'])
plt.legend()
plt.grid()
plt.show()
```

```python
# Sample dataset for energy consumption by sector over the years
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Residential': [1200, 1300, 1350, 1400, 1500],
    'Commercial': [800, 900, 950, 1000, 1100],
    'Industrial': [2000, 2000, 2000, 2100, 2200]
}

# Creating DataFrame
df_sectors = pd.DataFrame(data)

# Setting the positions and width for the bars
bar_width = 0.25
positions = range(len(df_sectors))

# Plotting grouped horizontal bars
plt.figure(figsize=(10, 6))
plt.barh([p - bar_width for p in positions], df_sectors['Residential'], bar_width, label='Residential', color='lightblue')
plt.barh(positions, df_sectors['Commercial'], bar_width, label='Commercial', color='lightgreen')
plt.barh([p + bar_width for p in positions], df_sectors['Industrial'], bar_width, label='Industrial', color='lightcoral')

# Adding labels and title
plt.yticks(positions, df_sectors['Year'])
plt.xlabel('Energy Consumption (in TWh)')
plt.title('Energy Consumption by Sector Over Years')
plt.legend()
plt.grid(axis='x')
plt.show()
```
